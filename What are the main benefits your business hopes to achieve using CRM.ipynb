```python
import matplotlib.pyplot as plt
import pandas as pd
data = {"services": ["Improve customer service", "Improve customer satisfaction","Improve customer retention", "Increase sales",
                     "Generate new business", "Ability to share and distribute data", "Market products & services more", "Improve market intelligence","Manage customer renewals", "Reduce costs", "All of these"],
        "percentage": [ 74, 66, 56, 54,53, 48, 37, 29, 24, 24, 8]}
df = pd.DataFrame.from_dict(data)
fig = plt.figure(figsize=(9, 5))
ax1 = plt.subplot(111)
df.sort_values("percentage", inplace=True)
#setting different bar color
ret = ax1.barh(df['services'], df['percentage'], color='#3399ff')
ret[6].set_color('#404040')
ret[10].set_color('#FF0000')
#================
plt.title("10 REASONS WHY SALES PEOPLE NEED CRM ~ by Zarema Plaksij ")
plt.xlabel("percentage(%)")
plt.ylabel("Main Business Benefits to achieve using CRM")
#================
ax1.xaxis.grid(linestyle = '--', linewidth=0.5)
for py, px in enumerate(df.percentage):
    if py ==10:
        ax1.annotate(px, xy=(px, py), fontstyle='italic', va='center', color= '#FF0000')
    else:
        ax1.annotate(px, xy=(px, py), fontstyle='italic', va='center')
ax1.set_xlim(0, df.percentage.max() * 1.1)
for pos in ['top', 'right', 'bottom', 'left']:
    ax1.spines[pos].set_linewidth(0.5)
    ax1.spines[pos].set_color('gray')
plt.show()                 
```
